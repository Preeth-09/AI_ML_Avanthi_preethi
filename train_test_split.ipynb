```python
# step 8 :prediction
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
#step9 : model metric calculation
```

```python
model.score(X_test,y_test)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('/content/archive (2).zip')
df
```

```python
X=df.drop('left',axis=1)
y=df['left']
```

```python
pd.crosstab(df.salary,df.left).plot(kind='bar')
```

```python
dummies=pd.get_dummies(df["Department"])
dummies
```

```python
dummies1=pd.get_dummies(df['salary'])
dummies1
```

```python
merge=pd.concat([X,dummies,dummies1],axis=1)
merge
```

```python
final=merge.drop(['salary','Department','low','technical'],axis=1)
final
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(final,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predicted = model.predict(X_test)
y_predicted
```

```python
model.score(X_test,y_test)
```
