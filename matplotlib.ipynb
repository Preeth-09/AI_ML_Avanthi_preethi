```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 9)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r' )
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = '#010200')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,2,4,4,6,8,10,12])
ypoints = np.array([1,2,1,4,3,5,1,6])

plt.plot(ypoints,'o-g', ms = 20, mec = 'g', mfc = '#ff0066',color='g',linestyle='dashed')
plt.show()
```
