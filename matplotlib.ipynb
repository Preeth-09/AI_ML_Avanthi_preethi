```python
import matplotlib.pyplot as plt
import numpy as np
plt.suptitle('CARTOONS',color='Red')
font1 = {'family':'serif','color':'blue','size':10}

x = np.array([0,1,2,3,4,5])
y = np.array([0,1,2,3,4,5])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('DOREMON', fontdict = font1)

x = np.array([0,1,2,3,4,5])
y = np.array([5,4,3,2,1,0])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title('SHINCHAN', fontdict = font1)

x = np.array([5,4,3,2,1,0])
y = np.array([0,1,2,3,4,5])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title('DORA', fontdict = font1)

x = np.array([0,1,2,3,4,5])
y = np.array([0,1,2,3,4,5])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title('MOTU PATLU', fontdict = font1)

plt.subplots_adjust(wspace=0.4,hspace=0.4)
plt.show()



```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint((50, size=(50))
y = np.random.randint((50, size=(50))
colors = np.random.randint(50, size=(50))
sizes = 40 * np.random.randint(50, size=(50))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
